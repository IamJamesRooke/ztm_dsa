Node Class

```python
class Node():
    def __init__(self, value):
        self.value = value
        self.next = None
```

Stack Class

```python
class Stack():
    def __init__(self):
        self.top = None
        self.bottom = None
        self.length = 0

    def peek(self):
        return self.top.value

    def push(self, value):
        
        newNode = Node(value)
        
        # if stack is empty
        if self.length == 0:    
            self.top = newNode
            self.bottom = newNode

        # if stack not empty
        else:
            holdingPointer = self.top
            self.top = newNode
            self.top.next = holdingPointer

        self.length += 1
        return self.show



    def pop(self):
        
        # if stack is empty:
        if self.length == 0:
            return None
        
        # if there's only one item in the stack
        if self.top == self.bottom:
            self.bottom = None
        
        holdingPointer = self.top
        self.top = self.top.next
        self.length -= 1

        return self.show
    


    @property
    def show(self):
        items = []
        current_node = self.top
        while current_node:
            items.append(current_node.value)
            current_node = current_node.next
        return items

    
myStack = Stack()
myStack.push('Discord')
myStack.push('Udemy')
myStack.push('Google')
myStack.push('Twitter')
myStack.peek()
```

Stack Array Implementation

```python
class Stack():
    def __init__(self):
        self.data = list()

    def peek(self):
        return self.data[0]

    def push(self, value):
        self.data.insert(0, value)

    def pop(self):
        return self.data.pop(0)

    @property
    def show(self):
        return self.data

    
myStack = Stack()
myStack.push('Discord')
myStack.push('Udemy')
myStack.push('Google')
myStack.push('Twitter')
myStack.peek()
```

Exercise: Queue Implementation

```python
import gc

class Node():
    def __init__(self, value):
        self.value = value
        self.next = None

class Queue():
    def __init__(self):
        self.first = None
        self.last = None
        self.length = 0

    def peek(self):
        return self.first.value

    def enqueue(self, value):

        newNode = Node(value)
        
        # if queue is empty
        if self.length == 0:
            self.last = newNode
            self.first = self.last
            self.length += 1
            return self

        # otherwise
        else:
            self.last.next = newNode
            self.last = newNode
            self.length += 1
            return self

    def dequeue(self):
        
        # if the queue is empty
        if self.length == 0:
            return 'empty queue'
        
        # otherwise...
        else:
            self.first = self.first.next
            self.length += 1
            return self

    def __repr__(self):
        queue = dict()
        current_node = self.first
        counter = 1
        while current_node:
            queue[counter] = current_node.value
            current_node = current_node.next
            counter += 1
        return str(queue)


myQueue = Queue()
myQueue.enqueue('John')
myQueue.enqueue('Adam')
myQueue.enqueue('Cindy')
myQueue.enqueue('David')
myQueue.dequeue()
myQueue.dequeue()
```
