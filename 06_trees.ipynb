# Trees

## Binary Trees

- **Level 0:** $2^0$ = 1 node
- **Level 1:** $2^1$ = 2 nodes (3 total)
- **Level 2:** $2^2$ = 4 nodes (7 total)
- **Level 3:** $2^3$ = 8 nodes (15 total)

### Binary Search Tree

**NOTE:** No $log N$ operations in binary search.

### Unbalanced Trees

Problem is that it can go too deep leading to $O(n)$ instead of $O(log N)$ time.



### Implementing a Binary Search Tree
```
        9
       / \
      4   20
     / \   / \
    1   6 15 170
```

````python
from graphviz import Digraph
from IPython.display import Image, display

class BinaryTreeNode():
    
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinarySearchTree():

    def __init__(self):
        self.root = None

    def insert(self, value):
        
        new_node = BinaryTreeNode(value)
        
        # if this is the first node of the tree
        # the new_node is assigned as the root
        if self.root is None:
            self.root = new_node
            return

        # otherwise, set the current node as the root
        current_node = self.root
        
        # keep going until the new_node is inserted
        while True:
            
            # check if the new node value is 
            # smaller than the current node value
            if value < current_node.value:
                
                # then check if the left child is empty
                if current_node.left is None:
                    
                    # if it's empty, the left child is the new ode
                    current_node.left = new_node
                    
                    # and we can exit the function
                    return
                
                # if the left isn't empty, the current node
                # will become the left child node
                # and we will loop back
                else:
                    current_node = current_node.left
            
            # same idea, if new node is >= current node
            else:
                # we check if the right node is empty
                if current_node.right is None:
                    
                    # if empty, the new node is now right child
                    current_node.right = new_node
                    
                    # and we exit the loop
                    return
                
                # otherwise the current node is now right child
                # and we loop back and try again
                else:
                    current_node = current_node.right

    def remove(self, value):
        pass

    def lookup(self, value):
        
        # first, check if empty
        if self.root == None:
            return False
        
        current_node = self.root

        while current_node:
            if value < current_node.value:
                current_node = current_node.left
            elif value > current_node.value:
                current_node = current_node.right
            elif value == current_node.value:
                return True
        return False

    def show_tree(self, node=None, graph=None, parent=None, direction=None):

        if self.root is None:
            print("The tree is empty.")
            return

        if graph is None:
            graph = Digraph(format='png')  # Ensure the format is set to PNG
            graph.attr(dpi='300')

        if node is None:
            node = self.root

        # Add the current node to the graph
        graph.node(str(node.value), str(node.value))

        # If there's a parent, add an edge
        if parent is not None:
            graph.edge(str(parent.value), str(node.value), label=direction)

        # Traverse left
        if node.left is not None:
            self.show_tree(node.left, graph, node, 'L')

        # Traverse right
        if node.right is not None:
            self.show_tree(node.right, graph, node, 'R')

        # If it's the root call, render the graph
        if parent is None:
            graph.render('binary_tree', format='png', cleanup=True)
            display(Image(filename='binary_tree.png'))


myTree = BinarySearchTree()
myTree.insert(9)
myTree.insert(4)
myTree.insert(20)
myTree.insert(1)
myTree.insert(16)
myTree.insert(170)
````
